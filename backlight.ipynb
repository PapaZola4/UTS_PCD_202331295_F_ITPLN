```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
backlight =cv2.imread('backlight.png')
```

```python
backlight.shape
```

```python
[baris, kolom] = backlight.shape[:2]
```

```python
rgb = cv2.cvtColor (backlight, cv2.COLOR_BGR2RGB)
```

```python
print(backlight.shape) 
print(backlight.dtype)
```

```python
b = backlight [:,:, 0]
g = backlight [:,:, 1]
r = backlight[:,:, 2]
```

```python
baris, kolom = backlight.shape[:2]
backlight_gray = np.zeros((baris, kolom))
for i in range(baris):
    for j in range(kolom):
        backlight_gray[i, j] = 0.2989 * r[i, j] + 0.587 * g[i, j] + 0.1141 * b[i, j]

        
backlight_gray = backlight_gray.astype(np.uint8)
cv2.imwrite('hasil_gray.jpg', backlight_gray)
fig, axs = plt.subplots(2,2, figsize =(15,5))
axs[0,0].imshow(backlight, cmap = 'gray')
axs[0,1].hist(backlight.ravel(), 256, [0,256])
axs[1,0].imshow(backlight_gray, cmap='gray')
axs[1,1].hist(backlight_gray.ravel(), 256, [0,256])
```

```python
alpha = 2.0
backlight_kontras = np.zeros((baris, kolom))
for i in range(baris):
    for j in range(kolom):
        gxy = alpha * backlight_gray[i, j]
        if gxy > 255:
            gxy = 255
        backlight_kontras[i, j] = gxy
backlight_kontras = backlight_kontras.astype(np.uint8)
cv2.imwrite('hasil_kontras.jpg', backlight_kontras)
fig, axs = plt.subplots(2,2, figsize =(15,5))
axs[0,0].imshow(backlight_gray, cmap= 'gray')
axs[0,1].hist(backlight_gray.ravel(), 256, [0,256])
axs[1,0].imshow(backlight_kontras, cmap='gray')
axs[1,1].hist(backlight_kontras.ravel(), 256, [0,256])
plt.show()
```

```python

beta =50   # kecerahan
backlight_cerah = np.zeros((baris, kolom))
for i in range(baris):
    for j in range(kolom):
        gxy =backlight_gray[i, j] + beta
        if gxy > 255:
            gxy = 255
        backlight_cerah[i, j] = gxy

backlight_cerah =  backlight_cerah.astype(np.uint8)
cv2.imwrite('backlight_cerah.jpg',  backlight_cerah)
fig, axs = plt.subplots(2,2, figsize =(15,5))
axs[0,0].imshow(backlight_gray, cmap= 'gray')
axs[0,1].hist(backlight_gray.ravel(), 256, [0,256])
axs[1,0].imshow( backlight_cerah, cmap='gray')
axs[1,1].hist( backlight_cerah.ravel(), 256, [0,256])
plt.show()
```

```python
alpha = 2.0  # kontras
beta = 3    # kecerahan
bright_contrast = np.zeros((baris, kolom))
for i in range(baris):
    for j in range(kolom):
        gxy = alpha * backlight_gray[i, j] + beta
        if gxy > 255:
            gxy = 255
        bright_contrast[i, j] = gxy

bright_contrast = bright_contrast.astype(np.uint8)
cv2.imwrite('bright_contrast.jpg', bright_contrast)
fig, axs = plt.subplots(2,2, figsize =(15,5))
axs[0,0].imshow(backlight_gray, cmap= 'gray')
axs[0,1].hist(backlight_gray.ravel(), 256, [0,256])
axs[1,0].imshow(bright_contrast, cmap='gray')
axs[1,1].hist(bright_contrast.ravel(), 256, [0,256])
plt.show()
```
